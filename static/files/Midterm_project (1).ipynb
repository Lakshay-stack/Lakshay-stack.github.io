U.S SUPERSTORE DATASET ANALYSIS

```python
# Loading  the dataset
import pandas as pd

# Reading the csv file
df = "/content/Sample - Superstore.csv"

# I've used encoding='ISO-8859-1' beacuse this file has special characters and python cannot read with default settings
store_df = pd.read_csv(df, encoding='ISO-8859-1')


# Displayed some rows

store_df.head()
```

Understanding Dataset Structure

```python
# dataset shape

print(store_df.shape)

# dataset columns
store_df.columns
```

Cleaning the data

```python
#checking any missing values

print(store_df.isnull().sum())

# droppig missing values (if any)
store_df = store_df.dropna()
# checking the data types
store_df.dtypes
```

Check Duplicates

```python
# checking the duplicates
print(store_df.duplicated().sum())

# Removing the duplicates

store_df = store_df.drop_duplicates
```

Insights & Analysis

```python
# Total Sales by Region

Total_sales = store_df.groupby('Region')['Sales'].sum().sort_values(ascending = False)
Total_sales

# The result shows us that the west region has the maximum sales whereas south has the least
```

```python
# Calculating profits according to sub-categories

profitable_subcategory = store_df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending= False)
profitable_subcategory

# It clearly shows that copiers has maximum profit whereas table category is in loss
```

```python
# Yearly Sales Trends

# Converting order date to datetime series
store_df['Order Date'] = pd.to_datetime(store_df['Order Date'])
#Extract Release year
store_df['Year'] = store_df['Order Date'].dt.year

# Calculating yearly sales trend

yearly_sales = store_df.groupby('Year')['Sales'].sum().sort_values(ascending = False)
yearly_sales

# The result shows that sales is increasing every year.
```

Filtering

```python
# Filtering all values that has sales > $1000

value_order = store_df.loc[store_df['Sales'] > 1000]
value_order


 # Top selling products
print(store_df['Product Name'].value_counts().head(10))
```
